```python
# Starter cell (optional)
import requests
import pandas as pd

API_URL = "https://remoteok.com/api"
```

```python
# Write your solution below.
# Tip: the API response includes a metadata object plus job objects.
# Make sure you keep only real job postings.

import requests
import pandas as pd

API_URL = "https://remoteok.com/api"
headers = {"User-Agent": "Mozilla/5.0"}

# Fetch data
response = requests.get(API_URL, headers=headers)
data = response.json()

# Filter real job postings
jobs = [job for job in data if isinstance(job, dict) and job.get("company")]

# Q1: How many job postings are currently available?
total_jobs = len(jobs)
print("Q1) Total Job Postings:", total_jobs)

# Q2: Top 3 most frequent companies
from collections import Counter
company_counts = Counter(job["company"] for job in jobs)
top3_companies = company_counts.most_common(3)
print("Q2) Top 3 Companies:", top3_companies)

# Q3: How many job titles contain "Data" (case-insensitive)?
count_data_titles = sum(1 for job in jobs if "data" in job.get("position", "").lower())
print('Q3) Job Titles with "Data":', count_data_titles)

# Q4: Create DataFrame of first 10 jobs
records = []
for job in jobs[:10]:
    title = job.get("position", "")
    company = job.get("company", "")
    location = job.get("location") or "Unknown"
    salary = job.get("salary") or job.get("salary_min", None)
    tags = ", ".join(job.get("tags", []))

    records.append({
        "title": title,
        "company": company,
        "location": location,
        "salary": salary,
        "tags": tags
    })

df = pd.DataFrame(records)
print("\nQ4) First 10 Job Postings DataFrame:")
print(df)











```

 Q2( 4 points)
Define a base class called 'Vehicle' with attributes make and model. Create a derived class Car that inherits from Vehicle and has an additional attribute 'num_doors'. Demonstrate an example of creating an instance of the 'Car' class and accessing its attributes.

```python
# Define base class
class Vehicle:
    def __init__(self, make, model):
        self.make = make
        self.model = model


# Define derived class
class Car(Vehicle):
    def __init__(self, make, model, num_doors):
        # Call parent constructor
        super().__init__(make, model)
        self.num_doors = num_doors


# Create an instance of Car
my_car = Car("Telsa", "Model S", 4)

# Access and print attributes
print("Make:", my_car.make)
print("Model:", my_car.model)
print("Number of Doors:", my_car.num_doors)
```

Question 3 ( 2 points)
Create a program that accepts a list of numbers as input and outputs a new list containing only the even numbers.

```python
# write your answer here
# Accept list of numbers from user
user_input = input("Enter numbers separated by spaces: ")

# Convert input string to list of integers
numbers = list(map(int, user_input.split()))

# Create a new list with only even numbers
even_numbers = [num for num in numbers if num % 2 == 0]

# Print result
print("Even numbers:", even_numbers)

```
