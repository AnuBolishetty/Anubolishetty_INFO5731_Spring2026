# **Question 1 (20 Points)**

**Dataset**: 20 Newsgroups dataset

**Dataset Link**: https://scikit-learn.org/0.19/datasets/twenty_newsgroups.html

**Consider Random 2000 rows only**

Generate K=10 topics by using LDA and LSA,
then calculate the coherence score and determine the optimal K value based on that score. Further, summarize and visualize each topic in your own words.


```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation, TruncatedSVD

# Load dataset (2000 samples)
dataset = fetch_20newsgroups(subset='all')
docs = dataset.data[:2000]

# Vectorization
count_vect = CountVectorizer(stop_words='english', max_df=0.95, min_df=2)
X_count = count_vect.fit_transform(docs)

tfidf_vect = TfidfVectorizer(stop_words='english', max_df=0.95, min_df=2)
X_tfidf = tfidf_vect.fit_transform(docs)

# LDA Model
lda = LatentDirichletAllocation(n_components=10, random_state=42)
lda.fit(X_count)

# LSA Model
lsa = TruncatedSVD(n_components=10, random_state=42)
lsa.fit(X_tfidf)

# Function to print topics
def print_topics(model, feature_names, n_top_words=10):
    for idx, topic in enumerate(model.components_):
        print(f"Topic {idx+1}:")
        print([feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]])
        print()

# Print topics
print("LDA Topics:")
print_topics(lda, count_vect.get_feature_names_out())

print("LSA Topics:")
print_topics(lsa, tfidf_vect.get_feature_names_out())

#summary: It concentrates on a preprocessing of the text data and the use of topic modeling methods to define effective patterns in the corpus. It often starts with cleaning the dataset(data lowercasing, removing punctuation, stopwords, and tokenization, etc.), then transforming the text into a numeric form, which could be CountVectorizer or TF-IDF Vectorizer. Once the features are extracted, the models such as LDA (Latent Dirichlet Allocation) or LSA (Latent Semantic Analysis) are used to find out the underlying topics in the text. Results come in the form of topic-word distributions, distribution of document-topic associations, which enable the interpretation of the key themes that are represented in the dataset. In general, the code illustrates the way in which raw text can be converted into structured insights by applying topic modeling.
```

```python
from gensim.models.coherencemodel import CoherenceModel
from gensim.corpora import Dictionary
import nltk
from bertopic import BERTopic
from sklearn.datasets import fetch_20newsgroups

nltk.download('punkt')
nltk.download('punkt_tab') # Download the missing punkt_tab resource

# Re-initialize and train BERTopic model since the previous cell was deleted
docs = fetch_20newsgroups(subset='all').data[:2000] # Ensure docs is available
topic_model = BERTopic(nr_topics=10) # Set nr_topics to aim for K=10, as requested in Q2
topics, probs = topic_model.fit_transform(docs)

texts = [nltk.word_tokenize(doc.lower()) for doc in docs]

dictionary = Dictionary(texts)

topics_words = []
for topic in topic_model.get_topics():
    if topic != -1:
        words = [word for word, _ in topic_model.get_topic(topic)]
        topics_words.append(words)

coherence_model = CoherenceModel(
    topics=topics_words,
    texts=texts,
    dictionary=dictionary,
    coherence='c_v'
)

print("Coherence Score:", coherence_model.get_coherence())
```
